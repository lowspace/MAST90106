```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Read company share price into df 1 to 18
df1 = pd.read_csv("./agriculture/YTEN.csv")
df2 = pd.read_csv("./agriculture/RKDA.csv")
df3 = pd.read_csv("./agriculture/CGA.csv") 
df4 = pd.read_csv("./agriculture/SEED.csv")
df5 = pd.read_csv("./agriculture/MBII.csv")


df6 = pd.read_csv("./agriculture/IPI.csv")
df7 = pd.read_csv("./agriculture/AVD.csv")
df8 = pd.read_csv("./agriculture/UAN.csv")
df9 = pd.read_csv("./agriculture/MGPI.csv")
df10 = pd.read_csv("./agriculture/ICL.csv")


df11 = pd.read_csv("./agriculture/CF.csv")
df12 = pd.read_csv("./agriculture/SMG.csv")
df13 = pd.read_csv("./agriculture/MOS.csv")
df14 = pd.read_csv("./agriculture/FMC.csv")
df15 = pd.read_csv("./agriculture/NTR.csv")


df16 = pd.read_csv("./agriculture/CTVA.csv")
df17 = pd.read_csv("./agriculture/CTA-PA.csv")
df18 = pd.read_csv("./agriculture/CTA-PB.csv")

                
```

```python
# Create Index list add company into one list
AgrIndex18List = [df1["Close"], df2["Close"], df3["Close"], df4["Close"], df5["Close"], df6["Close"]
                  , df7["Close"], df8["Close"], df9["Close"], df10["Close"], df11["Close"], df12["Close"],
                  df13["Close"], df14["Close"], df15["Close"], df16["Close"], df17["Close"], df18["Close"]]
                  
    
AgrIndex18List[1]

AgrIndex18List[1][1]
```

```python
### exclude the first company as the early stage share price too high over 70,000
### create agriculture index, adding share price together
AgrIndex = [0] * 15000
for i in range(1, 18):
    d = len(AgrIndex18List[i]) - 1
    n = 14999
    #print(i)
    while d > 0:
        #print(AgrIndex18List[i]["Close"][d])
        AgrIndex[n] = AgrIndex[n] + AgrIndex18List[i][d]
        #print(d)
        
        d -= 1
        n -= 1
```

```python
plt.plot(AgrIndex)
```

```python
# roughly year return

IndexR = []
r = 0
for i in range(len(AgrIndex) - 1):
    try:
        
        r = (AgrIndex[i + 350] - AgrIndex[i]) / AgrIndex[i]
        IndexR.append(r)
    except:
        pass
    
plt.plot(IndexR)
```
